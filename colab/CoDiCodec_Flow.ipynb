## CoDiCodec-Flow Usage Guide

This notebook demonstrates the basic workflow for using `CoDiCodec-Flow`. The steps involve **cloning the repository**, **setting up the environment**, **preprocessing data**, **training a model**, and **generating new audio**.

### 1. Setup
- We clone the repository and install the dependencies.
- Make sure to put your raw audio files in a folder like `./input_audio` before preprocessing.

### 2. Data Preprocessing
- We run the `preprocess` command to encode the raw audio files from `./input_audio` into latents.
- The processed data is saved in `./data/latents`.

### 3. Model Training
- We run the `train` command using the preprocessed data in `./data/latents`.
- Training logs and checkpoints are saved in `./runs/v0`.

### 4. Audio Generation
To generate audio after training, use a command like this:

```bash
!python cli.py generate \
    --model-path ./runs/v0/latest.ckpt \
    --out-dir ./generated_audio \
    --device cuda \
    --n-samples 5
```

````python
####### Preprocess

!python cli.py preprocess --in-dir ./input_audio \
  --out-dir ./data/latents                       \
  --device cuda
````

````python
####### Training

# !python cli.py train -h

!python cli.py train              \
    --data-dir   ./data/latents   \
    --out-dir    ./runs           \
    --device     cuda             \
    --batch-size 2                \
    --grad-accum 2                \
    --crop-tokens 512             \
    --max-steps 2000000           \
    --audio-unconditional         \
    --audio-sample-every 1000
````
